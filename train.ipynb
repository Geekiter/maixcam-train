# Generate .mud file

```python
model_name = "best_int8.cvimodel"
label = ["duck"]
save_path = "best_int8.mud"

# write to file
with open(save_path, 'w') as f:
    f.write("[basic]\n")
    f.write("type = cvimodel\n")
    f.write("model = %s\n" % model_name)
    f.write("\n")
    f.write("[extra]\n")
    f.write("model_type = yolov5\n")
    f.write("input_type = rgb\n")
    f.write("mean = 0, 0, 0\n")
    f.write("scale = 0.00392156862745098, 0.00392156862745098, 0.00392156862745098\n")
    f.write("anchors = 10,13, 16,30, 33,23, 30,61, 62,45, 59,119, 116,90, 156,198, 373,326\n")
    f.write("labels = %s\n" % ",".join(label))

print("done")
```
